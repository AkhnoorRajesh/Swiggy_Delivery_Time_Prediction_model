# Encoding

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
df.columns
```

```python
nominal=[ 'type_of_vehicle','festival','city_type','is_weekend']
ordinal=['weather', 'traffic']
numeric_cols=['ratings','vehicle_condition','multiple_deliveries','distance','order_time_hour','pickup_time_minutes']
```

```python
len(nominal)+len(ordinal)+len(numeric_cols)
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
pipe = Pipeline(steps=[('Nominal Encoding',OneHotEncoder()),
                       ('Ordinal Encoding',OrdinalEncoder()),
                       ('Feature Scaling',StandardScaler())])
```

```python
pipe
```

```python
Transformer=ColumnTransformer([('Nominal Encoding',OneHotEncoder(),nominal),
                       ('Ordinal Encoding',OrdinalEncoder(),ordinal),
                       ('Feature Scaling',StandardScaler(),numeric_cols)])
```

```python
df['traffic'].unique()
```

```python
seq_order=[['sunny','cloudy','windy', 'sandstorms', 'stormy'],['low','medium','high','jam']]
seq_order
```

```python
Transformer
```

```python
df.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = Pipeline([("Data Preprocessing",Transformer),
          ("ML Algorithm",LinearRegression())])
```

```python
X =df.drop('time_taken',axis = 1)
y = df['time_taken']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
model = Pipeline([("Data Preprocessing",Transformer),
          ("ML Algorithm",LinearRegression())])
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

# Model Evaluation

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
model_LR = Pipeline([("Data Preprocessing",Transformer),
          ("ML Algorithm",LinearRegression())])
model_LR.fit(X_train,y_train)
y_pred=model_LR.predict(X_test)
print("Linear Regression")
print("Mean Squared Error : ",mean_squared_error(y_test,y_pred))
print("R2 Score : ",r2_score(y_test,y_pred))
```

```python
df.columns
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model_RFR = Pipeline([("Data Preprocessing",Transformer),
          ("ML Algorithm",RandomForestRegressor())])
model_RFR.fit(X_train,y_train)
y_pred=model_RFR.predict(X_test)
print("RandomForestRegressor")
print("Mean Squared Error : ",mean_squared_error(y_test,y_pred))
print("R2 Score : ",r2_score(y_test,y_pred))
```
